```python
# date: 180315, AK
# import packages that are needed
import numpy as np
import scipy.io as sio
import pandas as pd 
import os
from os import listdir
from os.path import isfile, join
import csv
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
from matplotlib_scalebar.scalebar import ScaleBar
from IPython.display import display, HTML
from __future__ import division
```

```python
# load data from this directory
ctrl_timeline = np.load('C:/Users/annek/Documents/Lab/olfactory_bulb_hung/calcium/results/ctrl/ctrl_timeline_all.npy')
meflo_timeline = np.load('C:/Users/annek/Documents/Lab/olfactory_bulb_hung/calcium/results/mefloquine/mefloquine_timeline_all.npy')
# n-numbers
ns_ctrl = np.load('C:/Users/annek/Documents/Lab/olfactory_bulb_hung/calcium/results/ctrl/ctrl_n_numbers_all.npy')
ns_meflo = np.load('C:/Users/annek/Documents/Lab/olfactory_bulb_hung/calcium/results/mefloquine/mefloquine_n_numbers_all.npy')
```

```python
# save to this directory
save_plt_timeline = 'C:/Users/annek/Documents/Lab/olfactory_bulb_hung/calcium/results/ctrl_mefloquine_together/'
```

```python
# figure
# plot
fig = plt.figure()
plt.plot(ctrl_timeline, label='ctrl')
plt.plot(meflo_timeline, label='meflo')
plt.xlim(0, 30)
plt.ylim(0, 1.2)
plt.xlabel ('time (min)')
plt.ylabel ('$\Delta $ G/R (norm.)')
plt.legend()

# save
# png
fig.savefig(save_plt_timeline + 'timeline_ctrl_mefloquine.png')
# eps
plt.savefig((save_plt_timeline + 'timeline_ctrl_mefloquine.eps'), format='eps', dpi=1000)
# plt.close()
```

```python
# figure
# define time axis
time = np.array([2.5, 7.5, 12.5, 17.5, 22.5, 27.5])

# plot n-number
fig = plt.figure()
plt.plot(time, ns_ctrl, '-o', label='ctrl')
plt.plot(time, ns_meflo, '-o', label='meflo')
plt.legend()
plt.xlim(0, 30)
plt.ylim(0, 16)
plt.xlabel ('time (min)')
plt.ylabel ('n-number')

# save
# png
fig.savefig(save_plt_timeline + 'timeline_ctrl_mefloquine_n_number.png')
# eps
plt.savefig((save_plt_timeline + 'timeline_ctrl_mefloquine_n_number.eps'), format='eps', dpi=1000)
# plt.close()
```

```python
# figure
# ctrl and mefloquine and their n-numbers
time = np.array([2.5, 7.5, 12.5, 17.5, 22.5, 27.5])

# plot timeline
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax1.plot(ctrl_timeline, label='ctrl')
ax1.plot(meflo_timeline, label='meflo')
plt.xlim(0, 30)
plt.ylim(0, 1.4)
plt.xlabel ('time (min)')
plt.ylabel ('$\Delta $ G/R (norm.)')
plt.legend()
# plot n-number
ax1 = fig.add_subplot(2,1,2)
ax1.plot(time, ns_ctrl, '-o', label='ctrl')
ax1.plot(time, ns_meflo, '-o', label='meflo')
plt.xlim(0, 30)
plt.ylim(0, 16)
plt.xlabel ('time (min)')
plt.ylabel ('n-number')

# save
# png
fig.savefig(save_plt_timeline + 'timeline_overview.png')
# eps
plt.savefig((save_plt_timeline + 'timeline_overview.eps'), format='eps', dpi=1000)
# plt.close()
```
